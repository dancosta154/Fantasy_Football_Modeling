```python
import pandas as pd
import requests
import time
```

```python
weeks = [f'week-{week}' for week in range(1,17)]

for week in weeks:
    url = f'https://www.nflweather.com/en/week/2021/{week}/'
    res = requests.get(url)

res.status_code, res.text[:50]
```
