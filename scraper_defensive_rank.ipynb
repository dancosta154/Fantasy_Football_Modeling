```python
import pandas as pd
import requests
pd.options.display.max_columns = 999
```

### Scrape each week to get udpated defensive rankings.  Each week, export to csv and merge to master dataframe.

```python
week_num = 17 
#change number according to the week scrape is being done
```

```python
url = 'https://www.pro-football-reference.com/years/2021/opp.htm'
resp = requests.get(url)
resp

with open('./nfl_data/weekly/defensive_ranks.xls', 'wb') as f:
    f.write(resp.content)
```

```python
df_def = pd.read_html('./nfl_data/weekly/defensive_ranks.xls')
df_def = pd.DataFrame(df_def[0]) 
df_def= df_def.droplevel(0, axis=1)
df_def = df_def.rename(columns = {'Tm':'Team', 'Rk': 'Rank'})
df_def.head()
```

```python
# Grab only team name and rank
df_def = df_def.iloc[:, [0,1]].copy()
df_def.head()
```

```python
df_def.dtypes
```

```python
Team_Abbreviations_Dict = {
    'Arizona Cardinals': 'ARI',
    'Atlanta Falcons': 'ATL',
    'Baltimore Ravens': 'BAL',
    'Buffalo Bills' : 'BUF',
    'Carolina Panthers': 'CAR',
    'Chicago Bears': 'CHI',
    'Cincinnati Bengals': 'CIN',
    'Cleveland Browns': 'CLE',
    'Dallas Cowboys': 'DAL',
    'Denver Broncos': 'DEN',
    'Detroit Lions': 'DET',
    'Green Bay Packers': 'GNB',
    'Houston Texans': 'HOU',
    'Indianapolis Colts': 'IND',
    'Jacksonville Jaguars': 'JAX',
    'Kansas City Chiefs': 'KAN',
    'Miami Dolphins': 'MIA',
    'Minnesota Vikings': 'MIN',
    'New England Patriots': 'NE',
    'New Orleans Saints': 'NO',
    'New York Giants': 'NYG',
    'New York Jets': 'NYJ',
    'Las Vegas Raiders': 'LVR',
    'Philadelphia Eagles': 'PHI',
    'Pittsburgh Steelers': 'PIT',
    'Los Angeles Chargers': 'LAC',
    'San Francisco 49ers': 'SFO',
    'Seattle Seahawks': 'SEA',
    'Los Angeles Rams': 'LAR',
    'Tampa Bay Buccaneers': 'TAM',
    'Tennessee Titans': 'TEN',
    'Washington Commanders': 'WAS'
}
```

```python
# tack on column with corresponding team abbreviation
df_def['Tm_Abr'] = df_def['Team'].map(lambda x: Team_Abbreviations_Dict.get(x))
df_def
```

```python
df_def.drop(index = [32,33,34], axis = 0, inplace = True) #drop the average and total rows
```

```python
df_def['Week'] = week_num
df_def.head(10)
```

```python
df_def.to_csv(f'./nfl_data/weekly/defensive/defensive ranks for Week {week_num}.csv', index=0)
```
