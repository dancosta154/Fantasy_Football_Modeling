```python
import pandas as pd
import numpy as np
import requests
import os
import re

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
Team_Abbreviations_Dict = {
    'Arizona Cardinals': 'ARZ',
    'Atlanta Falcons': 'ATL',
    'Baltimore Ravens': 'BAL',
    'Buffalo Bills' : 'BUF',
    'Carolina Panthers': 'CAR',
    'Chicago Bears': 'CHI',
    'Cincinnati Bengals': 'CIN',
    'Cleveland Browns': 'CLE',
    'Dallas Cowboys': 'DAL',
    'Denver Broncos': 'DEN',
    'Detroit Lions': 'DET',
    'Green Bay Packers': 'GB',
    'Houston Texans': 'HOU',
    'Indianapolis Colts': 'IND',
    'Jacksonville Jaguars': 'JAX',
    'Kansas City Chiefs': 'KC',
    'Miami Dolphins': 'MIA',
    'Minnesota Vikings': 'MIN',
    'New England Patriots': 'NE',
    'New Orleans Saints': 'NO',
    'New York Giants': 'NYG',
    'New York Jets': 'NYJ',
    'Las Vegas Raiders': 'LV',
    'Philadelphia Eagles': 'PHI',
    'Pittsburgh Steelers': 'PIT',
    'Los Angeles Chargers': 'LAC',
    'San Francisco 49ers': 'SF',
    'Seattle Seahawks': 'SEA',
    'Los Angeles Rams': 'LAR',
    'Tampa Bay Buccaneers': 'TB',
    'Tennessee Titans': 'TEN',
    'Washington Commanders': 'WAS'
}
```

```python
file_list = []

for key, value in Team_Abbreviations_Dict.items():
    url = f'https://www.ourlads.com/nfldepthcharts/depthchart/{value}'
    res = requests.get(url)
    with open(f'./nfl_data/weekly/depth_chart/{key}.xls', 'wb') as f:
        f.write(res.content)
    file_list.append(f'./nfl_data/weekly/depth_chart/{key}.xls')
    data = pd.read_html(f'./nfl_data/weekly/depth_chart/{key}.xls')
    df = data[0]
    print(url)
```

```python
dfs = []
for file in file_list:
    data = pd.read_html(file)
    df = pd.DataFrame(data[0])
    df.drop(columns = ['No', 'No.', 'No.1', 'No.2', 'No.3',], inplace = True)
    df = df[df['Player 1'].apply(lambda x: len(str(x).split(',')) > 1)]
    df['Team'] = re.search(r'[^\/]+(?=\.[^\/.]*$)', file)[0].split()[-1]
    dfs.append(df)
df = pd.concat(dfs, ignore_index=True)
```

```python
# Clean dataframe
df['Player 1'] = [" ".join(str(i).title().split(' ')[::-1][1:]) for i in df['Player 1']]
df['Player 1'] = df['Player 1'].str.replace(',','')
df['Player 2'] = [" ".join(str(i).title().split(' ')[::-1][1:]) for i in df['Player 2']]
df['Player 2'] = df['Player 2'].str.replace(',','')
df['Player 3'] = [" ".join(str(i).title().split(' ')[::-1][1:]) for i in df['Player 3']]
df['Player 3'] = df['Player 3'].str.replace(',','')
df['Player 4'] = [" ".join(str(i).title().split(' ')[::-1][1:]) for i in df['Player 4']]
df['Player 4'] = df['Player 4'].str.replace(',','')
df['Player 5'] = [" ".join(str(i).title().split(' ')[::-1][1:]) for i in df['Player 5']]
df['Player 5'] = df['Player 5'].str.replace(',','')
```

```python
# Rename columns
df.rename(columns = {'Player 1': 'Depth_1',
                     'Player 2': 'Depth_2',
                    'Player 3': 'Depth_3',
                    'Player 4': 'Depth_4',
                    'Player 5': 'Depth_5',
                    }, inplace = True)
df.head()
```
