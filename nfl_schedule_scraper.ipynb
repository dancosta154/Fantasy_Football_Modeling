```python
import pandas as pd
import requests
import os
import re
import numpy as np

pd.options.display.max_columns = 999
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
if not os.path.exists('./nfl_data/weekly/schedule'):
    os.mkdir('./nfl_data/weekly/schedule')
```

```python
weeks = [week for week in range(1,18)]

file_list = []

for week in weeks:
    url = f'https://www.cbssports.com/nfl/schedule/2021/regular/{week}/'
    res = requests.get(url)
    print(res.status_code)
    output = open(f'./nfl_data/weekly/schedule/week{week}_schedule.xls', 'wb')
    output.write(res.content)
    file_list.append(f'./nfl_data/weekly/schedule/week{week}_schedule.xls')
    output.close()
    data = pd.read_html(f'./nfl_data/weekly/schedule/week{week}_schedule.xls')
    df = data
    print(f'printing week{week} schedule')
```

```python
file_list
```

```python
# get df started with the 1st file 'week-1_weather.xls'.  This way we have something to merge to.

data = pd.read_html(f'./nfl_data/weekly/schedule/week1_schedule.xls')
tnf = data[0]
sf = data[1]
mnf = data[2]
dfs = [tnf, sf, mnf]
schedule_df = pd.concat([df for df in dfs], ignore_index=True)
schedule_df['Week'] = re.search(r'(\d+)', file_list[0])[0]
schedule_df
```

```python
schedule_df.shape
```

```python
for file in file_list[1:16]: # Week 17 changed how they stored the data
        # print(file)
        data = pd.read_html(file)
        tnf = data[0]
        sf = data[1]
        mnf = data[2]
        dfs = [tnf, sf, mnf]
        df = pd.concat([df for df in dfs], ignore_index=True)
        df['Week'] = re.search(r'(\d+)', file)[0]
        dfs = [schedule_df, df]
        schedule_df = pd.concat([schedule_df for schedule_df in dfs], ignore_index=True)
schedule_df
```

```python
# Doing week 17 separately, because no TNF caused an issue -- so this is the easier route

data = pd.read_html(f'./nfl_data/weekly/schedule/week17_schedule.xls')
sf = data[0]
mnf = data[1]
dfs = [sf, mnf]
df = pd.concat([df for df in dfs], ignore_index=True)
df['Week'] = re.search(r'(\d+)', file_list[-1])[0]
dfs = [schedule_df, df]
schedule_df = pd.concat([schedule_df for schedule_df in dfs], ignore_index=True)
schedule_df['Week'] = schedule_df['Week'].astype(int)
schedule_df
```

```python
max_week = schedule_df['Week'].max()

schedule_df.to_csv(f'./nfl_data/weekly/schedule/schedule_through_week{max_week}.csv', index=0)
```
